```python
# Iterate through all the above countries with their respective ids and extract the fields listed below
# Save the information as a JSON labelled: <country name>_top_50.json
for index, i in enumerate(array):
    curr_charts =  sp.playlist_items(i, fields="items.track.album.id, items.track.type, items.track.album.id, items.track.album.name, items.track.album.release_date, items.track.artists.href, items.track.artists.id, items.track.artists.href, items.track.artists.name, items.track.artists.type, items.track.available_markets, items.track.duration_ms, items.track.id, items.track.name" , limit=50)
    country = country_array[index] + "_top50.json"
    with open(country, 'w', encoding='utf-8') as f:
            json.dump(curr_charts, f, ensure_ascii=False, indent=4)
            f.close()
```

```python
# Read each Json File and add important files/ primary keys accordingly
# Store all artists in dictionary as keys which will have song_ids as values
# Helps with creating "Sings" Table

dictionary_artist = defaultdict(set)
list_of_songs = []

for index, i in enumerate(country_array):
    curr_file = open(i+"_top50.json", "r")
    json_object = json.load(curr_file)
    curr_file.close()
    for index, curr_track in enumerate(json_object["items"]):
        json_object["items"][index]["track"]["country_chart"] = i
        json_object["items"][index]["track"]["chart_rank"] = index + 1
        json_object["items"][index]["track"]["album"]["song_id"] = json_object["items"][index]["track"]["id"]
        list_of_songs.append(json_object["items"][index]["track"]["album"]["song_id"])
        for curr_index, curr_artist in enumerate(json_object["items"][index]["track"]["artists"]):
            dictionary_artist[curr_artist["id"]].add(json_object["items"][index]["track"]["id"])
    list_of_songs = list(set(list_of_songs))
    print(i + ": " + str(len(list_of_songs)))
    new_file = open(i+"_top50.json", 'w', encoding='utf-8')
    json.dump(json_object, new_file, ensure_ascii=False, indent=4)
    new_file.close()
    
```

```python
# Create a dictionary that will contain all the genres that an artist is classified in
list_of_artists = list(dictionary_artist.keys())
dict_artist_genre = {}

for index in range(int(len(list_of_artists)/50) + 1):
    curr_list = list_of_artists[index * 50: (index + 1) * 50]
    curr_list_of_artists =  sp.artists(curr_list)
    for index in range(len(curr_list_of_artists["artists"])):
        dict_artist_genre[curr_list_of_artists["artists"][index]["id"]] = curr_list_of_artists["artists"][index]["genres"]
```

```python
# Create Song.csv
song_file = open("song.csv", "w+")
song_csv = csv.writer(song_file)
song_csv.writerow(["song_id", "name", "duration", "album_id", "country_chart", "chart_rank"])
song_file.close()
for i in country_array:
    curr_file = open(i + "_top50.json", "r")
    json_object = json.load(curr_file)
    curr_file.close()

    with open("song.csv", "a+") as song_file:
        song_csv = csv.writer(song_file)
        for x in json_object["items"]:
            song_csv.writerow([x["track"]["id"],
                            x["track"]["name"], 
                            x["track"]["duration_ms"],
                            x["track"]["album"]["id"],
                            x["track"]["country_chart"],
                            x["track"]["chart_rank"]])

```

```python
# Create album.csv
album_file = open("album.csv", "w+")
album_csv = csv.writer(album_file)
album_csv.writerow(["album_id", "album_name", "release_date"])
album_file.close()
for i in country_array:
    curr_file = open(i + "_top50.json", "r")
    json_object = json.load(curr_file)
    curr_file.close()

    with open("album.csv", "a+") as album_file:
        album_csv = csv.writer(album_file)
        for x in json_object["items"]:
            album_csv.writerow([x["track"]["album"]["id"],
                            x["track"]["album"]["name"], 
                            x["track"]["album"]["release_date"]])
```

```python
# Create sings.csv
# Relationship between songs.csv and artist.csv
sing_file = open("sing.csv", "w+")
sing_csv = csv.writer(sing_file)
sing_csv.writerow(["song_id", "artist_id"])
sing_file.close()
for i in country_array:
    curr_file = open(i + "_top50.json", "r")
    json_object = json.load(curr_file)
    curr_file.close()

    with open("sing.csv", "a+") as sing_file:
        sing_csv = csv.writer(sing_file)
        for x in json_object["items"]:
            for y in x["track"]["artists"]:
                sing_csv.writerow([x["track"]["id"], y["id"]])
```

```python
# Create artist.csv
artist_file = open("artist.csv", "w+")
artist_csv = csv.writer(artist_file)
artist_csv.writerow(["artist_id", "artist_name", "type"])
artist_file.close()
for i in country_array:
    curr_file = open(i + "_top50.json", "r")
    json_object = json.load(curr_file)
    curr_file.close()

    with open("artist.csv", "a+") as artist_file:
        artist_csv = csv.writer(artist_file)
        for x in json_object["items"]:
            for y in x["track"]["artists"]:
                artist_csv.writerow([y["id"],
                                     y["name"],
                                     y["type"]])
```

```python
# Create genre.csv
genre_file = open("genre.csv", "w+")
genre_csv = csv.writer(genre_file)
genre_csv.writerow(["artist_id", "genre"])
genre_file.close()
for i in country_array:
    curr_file = open(i + "_top50.json", "r")
    json_object = json.load(curr_file)
    curr_file.close()

    with open("genre.csv", "a+") as genre_file:
        genre_csv = csv.writer(genre_file)
        for x in json_object["items"]:
            for y in x["track"]["artists"]:
                get_genres = dict_artist_genre[y["id"]]
                for curr_genre in get_genres:
                    genre_csv.writerow([y["id"],
                                        curr_genre])
```

```python
# Create country.csv

country_file = open("country.csv", "w+")
country_csv = csv.writer(country_file)
country_csv.writerow(["country_code", "country_name"])
country_file.close()

list_of_countries = list(CC.keys())

with open("country.csv", "a+") as country_file:
    country_csv = csv.writer(country_file)
    for i in list_of_countries:
        country_csv.writerow([i,
                              CC[i]])
```

```python
# Create available_market.csv
available_market_file = open("available_market.csv", "w+")
available_market_csv = csv.writer(available_market_file)
available_market_csv.writerow(["album_id", "country_code"])
genre_file.close()
for i in country_array:
    curr_file = open(i + "_top50.json", "r")
    json_object = json.load(curr_file)
    curr_file.close()

    with open("available_market.csv", "a+") as available_market_file:
        available_market_csv = csv.writer(available_market_file)
        for x in json_object["items"]:
            for y in x["track"]["available_markets"]:
                available_market_csv.writerow([x["track"]["album"]["id"],
                                                y])
```
